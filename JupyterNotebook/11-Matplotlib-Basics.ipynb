# 11-Matplotlib-Basics

```python
import numpy as np 
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(np.sort(x), y[np.argsort(x)])
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

#### color parameterhttps://matplotlib.org/2.0.2/api/colors_api.html

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

#### linestyle parameterhttps://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) # x-range
plt.ylim(0, 1) # y-range
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2]) #first two is x range and last two is y range
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```

## Scatter plot

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```
